# Notes

- Considering X-axis for calculating saccadic velocity 
- Data points that increase by atleast 20 units in distance
- Time between each data point is 40 ms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data\Extracted-data\Saccades\Saccades-19.csv')

plt.plot(np.arange(0,df['LeftPupilPoints-X'].shape[0]),df['LeftPupilPoints-X'],color='red')
plt.plot(np.arange(0,df['LeftPupilPoints-Y'].shape[0]),df['LeftPupilPoints-Y'],color='blue')
plt.plot(np.arange(0,df['RightPupilPoints-X'].shape[0]),df['RightPupilPoints-X'],color='green')
plt.plot(np.arange(0,df['RightPupilPoints-Y'].shape[0]),df['RightPupilPoints-Y'],color='orange')

plt.title('Saccades-1')
plt.legend(['LeftPupilPoints-X','LefttPupilPoints-Y','RightPupilPoints-X','RightPupilPoints-Y'])
plt.show()
```

```python
count = 0
for i in df.index:
    if ((df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1]) > 20):
        count += 1
        print('Distance: ', df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1])
    elif ((df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1]) < -20):
        count += 1
        print('Distance: ', df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1])
print(count)
```

```python
count = 0
for i in df.index:
    if ((df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1]) > 20):
        count += 1
        print('Distance: ', df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1])
    elif ((df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1]) < -20):
        count += 1
        print('Distance: ', df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1])
print(count)
```

```python
count = 0
totalsum = 0
for i in df.index:
    
    if ((df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1]) > 10):
        totalsum += abs(df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1])
        count += 1
        
    elif ((df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1]) < -10):
        totalsum += abs(df['LeftPupilPoints-X'].iloc[i] - df['LeftPupilPoints-X'].iloc[i - 1])
        count += 1
```

```python
avg1 = (totalsum/count)/0.40
```

```python
count = 0
totalsum = 0
for i in df.index:
    
    if ((df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1]) > 10):
        totalsum += abs(df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1])
        count += 1
        
    elif ((df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1]) < -10):
        totalsum += abs(df['RightPupilPoints-X'].iloc[i] - df['RightPupilPoints-X'].iloc[i - 1])
        count += 1
```

```python
avg2 = (totalsum/count)/0.40
```

```python
print('Left saccadic velocity = ', avg1)
print('Right saccadic velocity = ', avg2)
```
