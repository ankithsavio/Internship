```python
a = df['LeftPupilPoints-X']
# a = df['RightPupilPoints-X']
```

```python
a = list(a)
b = list(a[1:])
```

```python
c =[]
for i,j in zip(a,b):
    c.append(j - i)
```

```python
a = c
b = c[1:]

d = []
for i,j in zip(a,b):
    d.append(j - i)
```

```python
count = 0
for i in d:
    if i > 15:
        print(i)
        count += 1
    elif i < -15:
        print(i)
        count+= 1
print('count = ' + str(count))
```

```python
a = df['LeftPupilPoints-X']
# a = df['RightPupilPoints-X']
```

```python
count = 0
pointy = []
pointx = []
for i in d:
    if i > 15:
        print(str(d.index(i)) + ' index  value: ' + str(a[d.index(i) + 1]) + ' end point')
        count += 1
        pointy.append(a[d.index(i) + 1])
        pointx.append(d.index(i) + 1)
    elif i < -15:
        print(str(d.index(i)) + ' index  value: ' + str(a[d.index(i) + 1]) + ' start point')
        count+= 1
        pointy.append(a[d.index(i) + 1])
        pointx.append(d.index(i) + 1)
print('count = ' + str(count))
```
